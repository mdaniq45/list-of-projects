```python
import pandas as pd

# Load with error handling
#df = pd.read_csv("paris.csv", error_bad_lines=False, warn_bad_lines=True)

# Inspect data
print(df.head())
print(df.info())
print(df.describe())

# Handle missing values
df = df.dropna()

# Fix inconsistent data
df['city'] = df['city'].str.strip().str.lower()

# Convert data types
df['date'] = pd.to_datetime(df['date'])

# Remove duplicates
df = df.drop_duplicates()

# Handle outliers
q_low = df['price'].quantile(0.01)
q_hi  = df['price'].quantile(0.99)
df = df[(df['price'] > q_low) & (df['price'] < q_hi)]

# Save cleaned data
df.to_csv("paris_cleaned.csv", index=False)
```
