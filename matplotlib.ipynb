```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

# Create a bar plot
plt.bar(categories, values)

# Add a title and labels
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python
data=[1,3,3,6,3,8,9,0,10]
plt.hist(data,bins=4)
plt.title('Histogram')
plt.xlabel('value')
plt.ylabel('Frequency')
plt.show()
```
