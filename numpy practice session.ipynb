```python
arr=np.array([2,4,6,8,10,12,14,16])
filter_arr=[]
for element in arr:
    if element %2==0:
        filter_arr.append(True)
    else:
        filter_arr.append(False)
        newarr=arr[filter_arr]
        print(filter_arr)
        print(newarr)
```
